```python
rooms = {1:{'Kitchen': 'yes' , 'Bathroom':'yes' , 'Hall' : 'yes'},
         2:{'Kitchen': 'no' , 'Bathroom':'no' , 'Hall' : 'no'},
         3:{'Kitchen': 'yes' , 'Bathroom':'no' , 'Hall' : 'no'}}
for i in rooms:
    print(i)
    for j in rooms[i]:
        print(j , ':' , rooms[i][j])
#What is the output?
#Ans :-
1
# Kitchen : yes
# Bathroom : yes
# Hall : yes
# 2
# Kitchen : no
# Bathroom : no
# Hall : no
# 3
# Kitchen : yes
# Bathroom : no
# Hall : no
```
