# Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
```

```python
os.chdir("C:\\Users\\Sai Prathap\\Downloads")
cars = pd.read_csv('Toyota.csv' , index_col = 0 , na_values = ['??' , '????'])
```

```python
cars.dropna(axis = 0 , inplace = True)
```

```python
plt.scatter(cars['Age'] , cars['Price'] , c = 'red')
```

```python
plt.title('Scatter plot of price vs age')
```

```python
plt.xlabel('Age(months)')
plt.ylabel('Price(Euros)')
```

```python
plt.scatter(cars['Age'] , cars['Price'] , c = 'red')
plt.title('Plot between age and price')
plt.xlabel('Age(months)')
plt.ylabel('Price(Euros)')
plt.show()
```

# Histogram

```python
plt.hist(cars['KM'])
```

```python
plt.hist(cars['KM'] ,color = 'green' , edgecolor = 'white',   bins = 5)
plt.title('Histogram of kilometers')
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```
