```python
#LINE GRAPH 
a = [1,2,3,4,5] #list1 
b = [1,2,8,3,4] #list2
c = (5,4,6,2,5) #tuple1

plt.plot(a,b,c = 'r', marker = "*", linewidth = 2, linestyle = "dashdot", label = "LINE 1")
plt.plot(a,c, c = 'k', marker = "^", linestyle = "dotted", label = "LINE 2")
plt.legend()
plt.show()  #IDLE, pycharm, vscode, spyder (use of plt.show() is compulsory in offline editors)
```

```python
#BAR GRAPH 
name = ['X','Y','Z','A']
height = [140,184,195,167]
plt.bar(name,height, color = ['k','r','b','gray'])
plt.xlabel("Name")
plt.ylabel("Height")
plt.title("Name vs Height comparision")
plt.show()
```

```python
#BAR GRAPH 
name = ['X','Y','Z','A']
height = [140,184,195,167]
plt.barh(name,height, color = ['k','r','b','gray'])
plt.xlabel("Height")
plt.ylabel("Name")
plt.title("Name vs Height comparision")
plt.show()
```

```python
#PIE CHART 
import matplotlib.pyplot as plt 
hours = ['2','9','9','4'] 
work = ['food','work/college','sleep','leisure']
plt.pie(hours, labels = work,startangle = 90, colors = ('r','green','c','#0920F0'), autopct = '%1.1f%%', explode = [0.5,0,0,0]) #gives the percentage from a range of 0 to 100
plt.legend(title = "ROUTINE")
plt.show()
```

```python
#SCATTER PLOTS
a = [1,2,3,4,5,6,7,8]
b = [5,4,7,3,2,9,1,3]
colors1 = ['r','b','g','k','magenta','cyan','purple','violet'] #list for colors 
plt.scatter(a,b, c = colors1, marker = ">", s = 150, alpha = 0.4) #alpha - transparency of points (0.0 to 1.0 )
```
